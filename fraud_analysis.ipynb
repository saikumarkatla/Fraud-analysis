```python
# drop name_orig and name_dest or extract useful information

df['orig_type']=df['name_orig'].str[0]
df['dest_type']=df['name_dest'].str[0]

df=df.drop(['name_orig','name_dest'],axis=1)
```

```python
df.head()
```

```python
# feature engineering
# transaction differences
df['balance_change_orig']=df['old_balance_org']-df['new_balance_orig']
df['balance_change_dest']=df['new_balance_dest']-df['old_balance_dest']
```

```python
df.head(2)
```

```python
# Error features
df['error_balance_orig']=df['new_balance_orig']+df['amount']-df['old_balance_org']
df['error_balance_dest']=df['old_balance_dest']+df['amount']-df['old_balance_dest']
```

```python
df.head(5)
```

```python
# extract hour from step
df['hour']=df['step']%24
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=['type'],drop_first=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['orig_type', 'dest_type'], drop_first=True)
```

```python
df.head()
```

```python
target='is_fraud'

features=['amount',
          'old_balance_org',
          'new_balance_orig',
          'old_balance_dest',
          'new_balance_dest',
          'is_flagged_fraud',
          'balance_change_orig',
          'balance_change_dest',
          'error_balance_orig',
          'error_balance_dest',
          'hour'
]
```

```python
# list comphrension
features+=[col for col in df.columns if col.startswith('type_')]
```

```python
features+=[col for col in df.columns if col.startswith('orig_type_')]
```

```python
features+=[col for col in df.columns if col.startswith('dest_type_')]
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='is_fraud', y='amount', data=df)
plt.yscale('log')  # log scale to handle skewed amounts
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[features]
y=df[target]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=3,random_state=3)
```

```python
from sklearn.ensemble import RandomForestClassifier
import time
```

```python
start=time.time()
model=RandomForestClassifier(n_estimators=100,class_weight='balanced',random_state=2,max_depth=15,n_jobs=-1)
model.fit(X_train,y_train)
end=time.time()

print(f'training time:{(end-start)/60:.2f}minutes')
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
y_prediction=model.predict(X_test)
```

```python
confusion_matrix(y_prediction,y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
confusion_matrix(y_prediction,y_test)
```

```python
classification_report(y_prediction,y_test)
```

```python
y_probability=model.predict_proba(X_test)[:,1]
```

```python
"ROC AUC Score:", roc_auc_score(y_test, y_probability)
```
